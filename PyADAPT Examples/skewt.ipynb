```python
from pyadapt.extras import skewt
import numpy as np

%matplotlib inline
```

```python
press = np.array([995,992,968,949,942,809])
temp = np.array([9.0,8.6,6.6,6.2,6.0,9.2])
height = np.array([229,254,455,610,678,1955])
u_wind = np.array([10,10,10,10,10,20])
v_wind = np.array([10,10,10,10,10,20])


fig, ax, bx = skewt.skewt_axes()

skewt.plot_profile(fig,ax,temp,press,label='T')

skewt.plot_wind(fig,bx,mask=(press>500),u=u_wind,v=v_wind,a=height,skip=1)
```
