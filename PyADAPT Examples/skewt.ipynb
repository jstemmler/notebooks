```python
from pyadapt.extras import skewt
import numpy as np

%matplotlib inline
```

```python
press = np.array([995,992,968,949,942,809])
temp = np.array([9.0,8.6,6.6,6.2,6.0,9.2])
dewp = np.array([0., 5., 3., 3., 3., -10.])
height = np.array([229,254,455,610,678,1955])
wind_dir = np.array([180, 210, 200, 200, 270, 270])
wind_spd = np.array([15,10,5,5,10,55])


fig, ax, bx = skewt.plot(pressure=press, altitude=height,
                         wind_dir=wind_dir, wind_spd=wind_spd,
                         profiles={'Temperature': temp, 'Dewpoint': dewp},
                         wind_interval=1)
```
